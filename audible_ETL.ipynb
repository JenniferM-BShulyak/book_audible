```python
import pandas as pd
import re
import datetime as dt
```

```python
df_clean = df = pd.read_csv("Resources/audible_cleaned.csv")
df_clean.head(10)
```

```python
# Import data 
df = pd.read_csv("Resources/audible_uncleaned.csv")
df.head(10)
```

```python
df_copy = df.copy()
```

### Cleaning the author and narrator

```python
# Write Function to clean author column
def clean_author(author):
    name= author.split(":")[1] # Split the string on the ":" and just take the name after the ":"
    res_list = re.findall('[A-Z][^A-Z]*', name) # Make a list by finding each name capitalized 
    return " ".join(res_list)   # return joined names separated by a space
```

```python
# Apply cleaning function to both author and narrator
df_copy["author"] = df_copy["author"].apply(lambda x: clean_author(x))
df_copy["narrator"]=df_copy["narrator"].apply(lambda x: clean_author(x))
df_copy.head()
```

### Clean the time

```python
# Function to convert the time into just minutes
def clean_time(time):
    time_list = re.findall('[0-9]+', time)
    if len(time_list) > 1:
        return int(time_list[0])*60 + int(time_list[1])
    return int(time_list[0])*60
    
```

```python
df_copy["time"] = df_copy["time"].apply(lambda x: clean_time(x))
df_copy.head()
```

### Clean release date

```python
def clean_date(d):
    return dt.datetime.strptime(d, "%d-%m-%y").date()
```

```python
df_copy["releasedate"] = df_copy["releasedate"].apply(lambda x: clean_date(x))
df_copy
```

### Add Year column

```python
df_copy.astype({'releasedate': 'datetime64'})
df_copy["release_year"] = df_copy['releasedate'].apply(lambda x: x.year)
df_copy.head()
```

### Clean stars

```python
def stars_reviews(stars_col):
    rating = []
    num_ratings = []
    for row in stars_col:
        if row == "Not rated yet":
            rating.append(row)
            num_ratings.append(0)
        else:
            r, five, n = re.findall(r'[-+]?\d*\.?\d+(?:,\d+)*|\d+', row)
            n_clean = n.replace(",", "")
            rating.append(r)
            num_ratings.append(n_clean)
    return rating, num_ratings
        
```

```python
df_copy["rating"], df_copy["number_of_ratings"]= stars_reviews(list(df_copy['stars']))
df_copy
```

```python
df_copy.drop(columns="stars", inplace=True)
df_copy
```
